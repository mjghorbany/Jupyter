```python
import csv,os
import pickle,re,sys
import numpy as np
import pandas
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from make_bow import make_baseline_feature_table
from make_bow import make_extended_ap_spreadsheet
from classify import just_dec_tree

from make_bow import rm_months_func
from make_bow import rm_yrs_func
from make_bow import rm_numbers_func
```

```python
def csv_to_df(csv_fp):
    """Load csv as dataframe with pandas. Make sure that
    data types are getting loaded correctly. Can add
    column names & types to dtype.
    """
    df = pandas.read_csv(csv_fp, dtype={'Account': np.float64})
    df = df[df['Account'].notnull()]
    return df
```

```python
accruals_csv = '../marritt_data/all-filtered_cleaned.csv'
ap_csv = "../marritt_data/2016 AP Query .xls"
merged_csv = '../marritt_data/merged.csv'

df_accruals = csv_to_df(accruals_csv)
len_rows_accruals = df_accruals.shape[0]  # 2445
df_accruals['accurals_index_col'] = df_accruals.index
df_ap = df = pandas.read_excel(ap_csv, sheetname='sheet1')
```

```python
df_marriott = pandas.read_csv(merged_csv, dtype={'Account': np.float64,'Invoice':object,'Descr.1':str,'Vendor':str})
df_marriott = df_marriott[df_marriott['Account'].notnull()]
```

```python
list_text_column = df_marriott['NewLineDescr'].tolist()
list_text_column = [rm_months_func(_str) for _str in list_text_column]
list_text_column = [rm_yrs_func(_str) for _str in list_text_column]
list_text_column = [rm_numbers_func(_str) for _str in list_text_column]
```

```python
print('Vectorizing')
vectorizer = TfidfVectorizer(analyzer='word',ngram_range=(1,3),lowercase=True)
print('Vectorizing0.5')
document_term_matrix = vectorizer.fit_transform(list_text_column)
document_term_matrix.toarray()
```
