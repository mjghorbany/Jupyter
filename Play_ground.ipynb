```python
import json
import re
from string import digits
dic_list=[]
wordcount = {}
file='/Users/jay.ghorbani/parse_defs/porto_dic.json'
with open(file) as fo:
        definitions_list = json.load(fo)
#print(definitions_list)
for entry in definitions_list:
    headword=entry['word']
    definition=entry['meaning']
    print('\n next')
    for word in definition.split():
        word = re.sub('([!@#$])', '', word)
        remove_digits = str.maketrans('', '', digits)
        word.translate(remove_digits)
        if word not in wordcount:
            wordcount[word] = 1
            #wordcount[index] = 1
        else:
            wordcount[word] += 1
            #wordcount[index] = 1
```
